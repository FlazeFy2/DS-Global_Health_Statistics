```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

df = pd.read_csv("Global Health Statistics.csv")

# List of disease
grouped_disease = df[['Disease Name']].drop_duplicates()
print(grouped_disease)
```

```python
# Exploratory Data Analysis (EDA) - Country with Most Population Affection of Disease
for _, dt in grouped_disease.iterrows():
    # Dataset
    data_disease = df[df['Disease Name'] == dt['Disease Name']]    
    top_country = data_disease.sort_values(by='Population Affected', ascending=False).head(5)
    top_country['Country_Year'] = top_country['Country'] + ' (' + top_country['Year'].astype(str) + ')'
    
    # Bar Chart
    fig, ax = plt.subplots()
    colors = cm.tab20.colors[:5]
    bars = ax.bar(top_country['Country_Year'], top_country['Population Affected'], color=colors)
    ax.set_ylabel('Population Affected')
    ax.set_title(f"Country with Most Population Affection of {dt['Disease Name']}")
    
    # Add value labels
    for bar in bars:
        height = bar.get_height()
        country_year = top_country['Country_Year'].values[0]
        ax.text(bar.get_x() + bar.get_width() / 2, height, 
                f'{height:.0f}', ha='center', va='bottom', fontsize=10)

    plt.xticks(rotation=25, ha='right')
    plt.tight_layout()
    plt.show()
```
