```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global Health Statistics.csv")

# Data Understanding - Variable & Data Type
print(df.dtypes)
print("\n")
print(f"{len(df.columns)} Column")
```

```python
# Data Preparation - Handle Missing Value
df.isnull().sum()
```

```python
# Data Understanding - Descriptive Statistic
df.describe()
```

```python
# Data Preparation - Outliers
ratio_column = ['Prevalence Rate (%)','Incidence Rate (%)','Mortality Rate (%)','Population Affected','Healthcare Access (%)','Doctors per 1000','Hospital Beds per 1000','Average Treatment Cost (USD)',
              'Recovery Rate (%)','DALYs','Improvement in 5 Years (%)','Per Capita Income (USD)','Education Index','Urbanization Rate (%)']

for dt in ratio_column:
    plt.boxplot(df[dt])
    plt.title(f'Boxplot of {dt}')
    plt.show()
```

```python
# Data Preparation - Invalid Value
ratio_column = ['Prevalence Rate (%)','Incidence Rate (%)','Mortality Rate (%)','Population Affected','Healthcare Access (%)','Doctors per 1000','Hospital Beds per 1000','Average Treatment Cost (USD)',
              'Recovery Rate (%)','DALYs','Improvement in 5 Years (%)','Per Capita Income (USD)','Education Index','Urbanization Rate (%)']
for dt in ratio_column:
    invalid_prices = df[df[dt] < 0]
    print(invalid_prices.index)
```
